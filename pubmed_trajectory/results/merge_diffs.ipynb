```python
import shutil
from pathlib import Path

PROJECT_ROOT = Path.cwd()
destination = PROJECT_ROOT / "results_2026_relaxed_diffs"

destination.mkdir(parents=True, exist_ok=True)

count = 0
for f in merged:
    try:
        with open(f, "r") as fp:
            data = json.load(fp)

        # skip files that are [] or empty
        if isinstance(data, list) and len(data) == 0:
            print(f"Skipping empty file: {f}")
            continue

        count += len(data)

        shutil.copy(f, destination)
        print(f"Copied: {f}")

    except json.JSONDecodeError:
        print(f"⚠️ Could not read JSON (skipped): {f}")
print(count)
```
